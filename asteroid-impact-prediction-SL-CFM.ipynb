```python
# Import Modules
import numpy as np 
import pandas as pd 
from pathlib import Path
```

```python
# Read in the orbits csv file into a pandas dataframe
orbits_df = pd.read_csv(Path("orbits.csv"))
orbits_df.head(7)
        
```

```python
# Number of rows
len(orbits_df.index)
```

```python
# Split "hazard" from the Object Classification column
orbits_df[["Object Classification", "Hazard"]] = orbits_df['Object Classification'].str.split('(', n=1, expand=True)
```

```python
orbits_df.head(7)
```

```python
# Strip the whitespace left on Object Classification column from the Hazard split
orbits_df["Object Classification"] = orbits_df["Object Classification"].str.strip()
```

```python
orbits_df.head(7)
```

```python
# Check for null values
orbits_df.isnull().sum()
```

```python
# Drop nulls
orbits_df.dropna(subset=["Object Name"], inplace=True)
```

```python
# Confirm dropped nulls
orbits_df.isnull().sum()
```

```python
# Drop random number 9 from Asteroid Magnitude column
orbits_df.dropna(subset=["Asteroid Magnitude"], inplace=True)
```

```python
orbits_df.isnull().sum()
```

```python
orbits_df.head(10)
```

```python
# Drop Object name as unnecessary
orbits_df.drop(['Object Name'],axis=1, inplace=True)
```

```python
# Confirm dropped column
orbits_df.head()
```

```python
# Use get dummies to convert Hazard column to binary values
hazard_type = pd.get_dummies(orbits_df['Hazard'], dtype=int)
hazard_type
```

```python
# Check counts for balance of classes
hazard_type.value_counts()
```

```python
# Concatenate original df with hazard df created with get dummies
orbits2_df = pd.concat([orbits_df, hazard_type], axis=1)
orbits2_df.head()
```

```python
# Rename hazard_type column to prepare for dropping original column
orbits3_df = orbits2_df.rename(columns={"Hazard)":"Hazardous"})
orbits3_df.head()
```

```python
# Drop original Hazard column
orbits3_df = orbits3_df.drop(["Hazard"], axis=1)
orbits3_df.tail()
```

```python
# Use get dummies to convert Object classification to binary values
obj_class_df = pd.get_dummies(orbits_df['Object Classification'], dtype=int)
obj_class_df
```

```python
#Concatenate orbits3_df with obj_class_df created with getdummies
orbits4_df = pd.concat([orbits3_df, obj_class_df], axis=1)
orbits4_df.head()
```

```python
# Drop original Object Classification column
orbits4_df = orbits4_df.drop(["Object Classification"], axis=1)
orbits4_df.tail()
```

```python
# Shape for heatmap
orbits4_df.shape
```

```python
# Import dependencies for correlation matrix
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import seaborn as sns
```

```python
# Heatmap of correlation
plt.figure(figsize = (18,18))
sns.heatmap(orbits4_df.corr(),annot = True)
```

```python
 # Get the target variables. 
target = orbits4_df["Hazardous"]
target_names = ["safe", "hazard"]
```

```python
 # Get the features. 
data = orbits4_df.drop("Hazardous", axis=1)
feature_names = data.columns
data.head()
```

Data is ready for the Supervised Learning process. Begin building the model.

```python
 # Split data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

Using StandardScaler to transform data to determine in effort to get better results.

```python
 # Creating StandardScaler instance
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
 # Fitting Standard Scaler
X_scaler = scaler.fit(X_train)
```

```python
 # Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# View scaled data
X_train_scaled
```

```python
 # Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
 # Calculate the classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
# Low precision and recall due to imbalance of data classes (cell 14, 0:13855, 1:1779)
```

```python
# scikit learn confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predictions)
cm
```

```python
# Confusion Matrix with labels (https://medium.com/@dtuk81/confusion-matrix-visualization-fc31e3f30fea)
group_names = ['True Safe','False Safe','False Hazard','True Hazard']
group_counts = ["{0:0.0f}".format(value) for value in
                cm.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cm.flatten()/np.sum(cm)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cm, annot=labels, fmt='', cmap='Blues')
```
