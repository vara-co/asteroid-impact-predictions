## Building the ML model

```python
# Import the modules for the ML
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

```python
# Make a copy of the DataFrame
orbit_df_copy = orbit_df.copy()
```

```python
# Use get dummies to convert Object classification to binary values
obj_class_df = pd.get_dummies(orbit_df['Object Classification'], dtype=int)
obj_class_df
```

```python
# Use get dummies to convert Object classification to binary values
hazard_df = pd.get_dummies(orbit_df['Hazardous'], dtype=int)
hazard_df
```

```python
# Concatenate Original Dataframe with the Hazard dataframe

orbits_concat = pd.concat([orbit_df_copy, hazard_df], axis=1)
orbits_concat.head()
```

```python
# Concatenate Previous Dataframe with the Object Classification dataframe

orbits_concat_objclass_haz = pd.concat([obj_class_df, orbits_concat], axis=1)
orbits_concat_objclass_haz.head()
```

```python
# IF above we were dropping more than one column
orbits_concat2 = orbits_concat_objclass_haz.drop(columns=['Object Classification', 'Hazardous'])

# Show sample DataFrame
orbits_concat2.head()
```

```python
# Split preprocessed data into features and target arrays
# Define features set
# Make a copy of the Dataset
X = orbits_concat2.copy()

# Drop the column
X.drop(columns=['Hazard'], axis=1, inplace=True)

# Display the values the features columns
X.head()
```

```python
# Define Target Vector
y = orbits_concat2['Hazard'].values

# Show first 5 values in the array
y[:5]
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
# Transform the training data
X_train_scaled = X_scaler.transform(X_train)

# Transform the testing data
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.activations import relu
```

```python
# Define a learning rate scheduler
def scheduler(epoch, lr):
    if epoch < 10:
        return lr
    else:
        return lr * tf.math.exp(-0.1)
```

```python
# Define the model
nn = tf.keras.models.Sequential()

# Input layer with swish activation and L2 regularization
nn.add(Dense(units=40, kernel_regularizer=l2(0.001), input_dim=X_train_scaled.shape[1], activation='relu'))
nn.add(BatchNormalization())
nn.add(Dropout(0.2))

# Additional hidden layers with relu and L2 regularization
nn.add(Dense(units=20, activation='relu', kernel_regularizer=l2(0.001)))
nn.add(BatchNormalization())
nn.add(Dropout(0.2))

# Additional hidden layers with relu and L2 regularization
nn.add(Dense(units=20, activation='relu', kernel_regularizer=l2(0.001)))
nn.add(BatchNormalization())
nn.add(Dropout(0.2))

# Additional hidden layers with relu and L2 regularization
##nn.add(Dense(units=20, activation='relu', kernel_regularizer=l2(0.001)))
##nn.add(BatchNormalization())
##nn.add(Dropout(0.2))

# Output layer
nn.add(Dense(units=1, activation='sigmoid'))

nn.summary()
```

```python
# Compile the model
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Define callbacks
checkpoint_callback = ModelCheckpoint(filepath='checkpoint_weights.h5', save_weights_only=True, save_freq='epoch')
early_stopping = EarlyStopping(monitor='val_loss', patience=10)
lr_scheduler = LearningRateScheduler(scheduler)
```

```python
# Train the model and use the callbacks so that we can stop  the validation if the training stops improving
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, validation_split=0.15, validation_data=(X_test_scaled, y_test), callbacks=[checkpoint_callback, early_stopping, lr_scheduler])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
